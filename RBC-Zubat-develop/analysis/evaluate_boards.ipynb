```python
import re

import os
os.environ["STOCKFISH_EXECUTABLE"] = "/home/jacek/Documents/uni/masters/stockfish/stockfish-ubuntu-x86-64"

import pandas as pd
from reconchess import GameHistory, WinReason
import numpy as np
import json
from tqdm import tqdm
from chess.engine import Limit
import chess
```

```python
from strangefish.utilities.stockfish import create_engine

engine = create_engine()
```

```python
def add_scores(path, game_id):
    df = pd.read_csv(os.path.join(path, f'move_scores_{game_id}.csv'))
    # if 'true_score' in df.columns:
    #     return
    
    game = GameHistory.from_file(os.path.join(path, f'game_{game_id}.log'))
    results = []
    color = True if game.get_white_player_name() == 'Zubat' else False

    for turn in tqdm(game.turns(color)):
        fen = game.truth_fen_before_move(turn)
        board = chess.Board(fen)
        if board.is_valid():
            res = engine.analyse(board, limit=Limit(time=0.05, depth=10))
            try:
                results.append(res['score'].relative.cp)
            except:
                results.append(8000 * (1 if res['score'].relative.moves > 0 else -1))
        else:
            results.append(8000 * (1 if color else -1))

    df['true_score'] = [results[i-1] for i in df['move_number']]
    df.to_csv(os.path.join(path, f'move_scores_{game_id}.csv'))
```

```python
base_path = 'game_logs/unranked_games'
for group in sorted(os.listdir(base_path)):
    print(group)
    for bot in sorted(os.listdir(os.path.join(base_path, group))):
        print(f'{group}-{bot}')
        for game_id in tqdm(sorted(os.listdir(os.path.join(base_path, group, bot)))):
            try:
                game_path = os.path.join(base_path, group, bot, game_id)
                add_scores(game_path, game_id)
            except Exception as e:
                print(e)
```
